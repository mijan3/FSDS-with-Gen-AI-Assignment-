### 2. Count Word Frequency in Customer Reviews 

 Real-world use: Analyzing reviews or comments in text analytics. 

Task: 

Given a review text, count how many times each word appears (ignore case). 

Example: 

Input: "This product is good and this service is excellent" 

Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
review=input("Enter your review: ").lower()
words=review.split()
feq={}
for word in words:
    if word in feq:
        feq[word]+=1
    else:
        feq[word]=1 
print(feq)
```

### 3. Password Strength Checker 

Real-world use: Used in user authentication systems.

Task: 

Check if a password is strong — must have at least one uppercase, one lowercase, one digit, and one 
special character.

Example: 
Input: "Hello@123" 
Output: Strong Password

```python
password = input("Enter your password: ")

has_upper = False
has_lower = False
has_digit = False
has_special = False
special_characters = "!@#$%^&*()-+=_{}[]|;:'\",.<>?/`~"
for char in password:
    if 'A'<=char<='Z':
        has_upper = True
    elif 'a'<=char<='z':
        has_lower = True
    elif '0'<=char<='9':
        has_digit = True
    elif char in special_characters:
        has_special = True 
if has_upper and has_lower and has_digit and has_special:
    print("Strong Password")
else:
    print("Weak Password")
```

### 4. Username Generator 

Real-world use: Automatically create usernames for users.

Task: 
Take user’s full name and generate a short username. 

Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only). 

Example: 
Input: "Mamun Malitha" 
Output: mamitha57 

```python
import random
full_name = input("Enter your full name: ")
first_name=""
last_name=""
space_found=False
for char in full_name:
    if char==" ":
        space_found=True
        continue
    if not space_found:
        first_name+=char
    else:
        last_name+=char
username = first_name[:3].lower() + last_name[-3:].lower()
random_number = ""
for _ in range(2):
    random_number += str(random.randint(0, 9))  
username += random_number
print(username)
```

### 5. SMS Word Counter 

Real-world use: Used in messaging apps or social media post limit check.

Task: 
Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).

Example: 

Input: "This is a demo message." 

Output: Words = 5, Characters = 23, Status = Within Limit 

```python
message=input("Enter your message: ")

char_count=0
for char in message:
    char_count+=1

word_count=0
in_word=False
for char in message:
    if char.isalpha():
        if not in_word:
            word_count+=1
            in_word=True
    else:
        in_word=False
        
if char_count>160:
    status="Exceeds limit"
else:
    status="Within limit"
print(f"Words= {word_count},Characters= {char_count}, Status={status}")
```

### 6. Remove Duplicates from Contact List 

Real-world use: Data cleaning in CRM or user databases. 

Task: 
Given a list of names (comma separated), remove duplicates. 

Example: 

Input: "Rahim, Karim, Rahim, Sultana" 

Output: "Rahim, Karim, Sultana"

```python
contacts_input = input("Enter contact names (comma separated): ")

name=""
names=[]
unique_names=[]

for char in contacts_input:
    if char==",":
        names.append(name.strip())
        name=""
    else:
        name+=char
if name:
    names.append(name.strip())
for n in names:
    found=False
    for un in unique_names:
        if n.lower()==un.lower():
            found=True
            break
    if not found:
        unique_names.append(n)
        
result=''
for i in range(len(unique_names)):
    result+=unique_names[i]
    if i!=len(unique_names)-1:
        result+=", "
print(result)
```

### 7. Detect Suspicious Words in Text (Moderation) 

 Real-world use: Used in chat moderation or email filtering. 

Task: 
If any “banned” words like ['spam', 'scam', 'fake'] are found in a text, print “Warning”.
 
Example: 

Input: "This is a scam offer"
 
Output: Warning: Suspicious content detected 

```python
text=input("Enter your text: ")

banned_words = ['spam', 'scam', 'fake']
word=""
words=[]
for char in text:
    if char==" ":
        words.append(word)
        word=""
    else:
        word+=char
if word:
    words.append(word)
    
worning=False
for w in words:
    for bw in banned_words:
        if w.lower()==bw:
            worning=True
            break
    if worning:
        break
if worning:
    print("Warning: Suspicious content detected ")
else:
    print("Content is clean.")
```

8. Invoice Code Formatter

Real-world use: Formatting IDs or codes in finance or ERP systems. 

Task: 
Given an invoice number like 23, generate code in the format INV00023. 

Example: 

Input: 23 

Output: INV00023

```python
num=int(input("Enter invoice number: "))

num_str=''
if num==0:
    num_str='0'
else:
    while num>0:
        digit=num%10
        num_str=str(digit)+num_str
        num=num//10
count=0
for char in num_str:
    count+=1
    
while count<5:
    num_str='0'+num_str
    count+=1
invoice_code="INV"+num_str
print(invoice_code)
```

### 9. Count Hashtags and Mentions (Social Media Analytics) 

Real-world use: Used in analyzing social media posts. 

Task: 
Count how many hashtags (#) and mentions (@) appear in a post. 

Example: 

Input: "Check out #AIWorkshop and follow @inceptionbd" 

Output: Hashtags = 1, Mentions = 1

```python
post=input("Enter your social media post: ")
hashtag_count=0
mention_count=0

for char in post:
    if char=="#":
        hashtag_count+=1
    elif char=="@":
        mention_count+=1
        
print(f"Hashtags= {hashtag_count}, Mentions= {mention_count}")
```

### 10. Word Reversal for Data Encryption 

 Real-world use: Simple text obfuscation or playful encoding.

Task: 
Reverse each word in a sentence while keeping word order the same. 

Example: 

Input: "AI is fun" 

Output: "IA si nuf"

```python
text=input("Enter your text: ")

word=""
reversed_text=""
for char in text:
    if char==" ":
        rev=""
        for c in word:
            rev=c+rev
        reversed_text+=rev+" "
        word=""
    else:
        word+=char
        
rev=""
for c in word:
    rev=c+rev
reversed_text+=rev

print(reversed_text)
```
